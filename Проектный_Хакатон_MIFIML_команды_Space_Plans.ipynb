### Бизнес-постановка задачи

* **Проблема**

  * Окружающая среда постоянно меняется под воздействием природных факторов и деятельности человека. Экологам, научным работникам и представителям власти необходимы достоверные данные для мониторинга состояния экосистем, принятия обоснованных решений по сохранению и реагированию на экологические проблемы. Однако традиционные методы сбора данных трудоемки и часто показывают высокий процент ошибок.

  * Точные и оперативные экологические исследования приобретают все большее значение в связи с усилением антропогенного воздействия и изменением климата. Мониторинг флоры и фауны, качества воздуха, воды и состояния земель требует интеграции различных типов и форматов данных.

* **Цель**
  1. Интеграция предварительно обученных моделей машинного обучения для анализа данных о загрязнении воздуха, чтобы:
    * Прогнозировать общий индекс качества воздуха (AQI ).
    * Классифицировать уровень загрязнения (AQI Category).
    * Оценивать влияние отдельных загрязнителей (CO, Ozone, NO2, PM2.5) на качество воздуха.

* **Этапы реализации**
  1. **Подготовка данных:**
    * Загрузка и очистка табличных данных, содержащих информацию о стране, городе, уровне загрязнения и категории AQI.
    * Кодирование категориальных данных (Country, City, AQI Category) с использованием методов, таких как Label Encoding.

 2. **Обучение модели:**
    * Использование предварительно обученных моделей CatBoost для выполнения следующих задач:
      * Прогнозирование значений AQI Value.
      * Классификация AQI Category.
      * Оценка влияния загрязняющих веществ на AQI.
      * Применение методов машинного обучения для обучения модели с использованием показателей процентного содержания загрязняющих веществ и географическое положение.
  
  3. **Оценка модели:**
    * Оценка качества модели с помощью различных метрик:
    * Для задачи регрессии: Mean Squared Error (MSE), R² Score.
    * Для задачи классификации: Accuracy
    * Проведение кросс-валидации для повышения надежности результатов.
    
  4. **Визуализация и интерпретация результатов:**
    * Построение графиков важности признаков, чтобы понять, какие вредные вещества оказывают наибольшее влияние на результаты.
    * Визуализация предсказаний модели и реальных значений AQI.

Корреляция O₃ и PM2.5 (0,26). Оба зягрязнителя O₃ и PM2.5 имеют имеют общие прекурсоры - оксиды азота. Также озон реагирует с другими загрязнителями (например, аммиаком, серой), усиливая образование вторичных частиц.



## Преобразование данных

```python
df = data.copy()
```

```python
columns_to_convert = [
    "PM2.5 AQI Category",
    "NO2 AQI Category",
    "Ozone AQI Category",
    "CO AQI Category",
    "AQI Category"
]

# columns на 'category'
df[columns_to_convert] = df[columns_to_convert].astype('category')

# Тест
print(df.dtypes)

# Проверка, что все пропуски были заменены
print(df.isnull().sum())
```

```python
# Создаем словарь для последующую сравнения с прогнозом
aqi_category_mapping = dict(enumerate(df['AQI Category'].astype('category').cat.categories))

aqi_category_mapping
```

```python
# Кодирование категориальных данных (Label Encoding)
df['AQI Category'] = df['AQI Category'].astype('category').cat.codes

# Преобразование категорий загрязнителей
for col in ['CO AQI Category', 'Ozone AQI Category', 'NO2 AQI Category', 'PM2.5 AQI Category']:
    if col in df.columns:
        df[col] = df[col].astype('category').cat.codes


# Выбор фич и целевой переменной
X = df[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']]
y_value = df['AQI Value']  # Для прогноза AQI
y_category = df['AQI Category']  # Для классификации AQI Category

# Разделение на тренировочные и тестовые данные
X_train, X_test, y_train, y_test = train_test_split(X, y_value, test_size=0.2, random_state=42)
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_category, test_size=0.2, random_state=42)
```

## Обучение модели

### Прогнозирование AQI (регрессия)

```python
# Первичная инициализация модели CatBoost для регрессии
model_reg = CatBoostRegressor(iterations=10, depth=1, learning_rate=0.1, loss_function='RMSE')
```

```python
%%time
##Поиском по сетке подбираем самые лучше параметры по модели регрессии
# Время выполнения ~5 минут!
param_grid_reg = {
        'iterations': [100, 500, 1000],
        'depth': [2, 4, 6],
        'learning_rate': [0.001, 0.01, 0.1]
    }
gs_reg = GridSearchCV(model_reg, param_grid_reg, scoring='neg_mean_squared_error', cv=5)
gs_reg.fit(X_train, y_train)
best_params_reg = gs_reg.best_params_
best_score_reg = (-gs_reg.best_score_) ** 0.5  # Преобразование из отрицательной MSE в RMSE
print(f"Лучшие параметры для model_reg: {best_params_reg}")
print(f"RMSE для model_reg: {best_score_reg}")
```

```python
%%time
#Обучаем модель с лучшими параметрами
model_reg = CatBoostRegressor(iterations=best_params_reg['iterations'], depth=best_params_reg['depth'], learning_rate=best_params_reg['learning_rate'], loss_function='RMSE')

model_reg.fit(X_train, y_train)

# Оценка модели
y_pred_reg = model_reg.predict(X_test)
```

### Классификация AQI Category

```python
# Инициализация создание модели CatBoost для классификации
# Параметр loss_function='MultiClass' указывает, что модель будет обучаться для предсказания нескольких классов (многоклассовая классификация).
#Это означает, что целевая переменная (y_train_class) содержит более двух уникальных классов, и модель будет предсказывать вероятность принадлежности к каждому из этих классов.
model_class = CatBoostClassifier(iterations=10, depth=1, learning_rate=0.1, loss_function='MultiClass')
```

```python
%%time
##Поиском по сетке подбираем самые лучше параметры для модели классификации
# Время выполнения от 10 минут!
param_grid_class = {
        'iterations': [100, 500, 1000],
        'depth': [2, 4, 6],
        'learning_rate': [0.001, 0.01, 0.1]
    }
gs_class = GridSearchCV(model_class, param_grid_class, scoring='accuracy', cv=5)
gs_class.fit(X_train_class, y_train_class)
best_params_class = gs_class.best_params_
best_score_class = gs_class.best_score_
print(f"Лучшие параметры для model_reg: {best_params_class}")
print(f"accuracy для model_class: {best_score_class}")
```

```python
%%time
#Обучаем модель с лучшими параметрами
model_class = CatBoostClassifier(iterations=best_params_class['iterations'], depth=best_params_class['depth'], learning_rate=best_params_class['learning_rate'], loss_function='MultiClass')

# Обучение модели
model_class.fit(X_train_class, y_train_class)

# Оценка модели
y_pred_class = model_class.predict(X_test_class)
```

```python
# Получение важности признаков
feature_importances = model_class.get_feature_importance(Pool(X_train_class, y_train_class))

features = X.columns

# Построение графика
plt.figure(figsize=(10, 6))
with sns.color_palette("flare"):
    sns.barplot(x=feature_importances, y=features, hue=features)
plt.title("Оценка важности признаков в CatBoost Model")
plt.xlabel("Оценка важности")
plt.ylabel("Признак (вх.параметр)")
plt.show()
```

```python
#Используем SHAP чтобы наглядно показать два главных влияющих параметра на AQI
explainer = shap.TreeExplainer(model_reg)
shap_values = explainer.shap_values(X_test)

shap.summary_plot(shap_values, X_test)
```

## Оценка модели

```python
# Оценка качества модели для прогноза AQI
mse = mean_squared_error(y_test, y_pred_reg)
r2 = r2_score(y_test, y_pred_reg)

print(f'Mean Squared Error: {mse}')
print(f'R2 Score: {r2}')
```
